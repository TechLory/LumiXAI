```python
from client import Client

client = Client()
client.clear_history()
```

```python

PROMPTS = [
    "a painting of a cow in the Rembrandt style",
    "a painting of a cow in the Impressionism style",
    "a painting of a dog in the Baroque style",
    "a painting of a person in the Cubism style",
]

batch = [
    {
        "source": "huggingface",
        "model": "stabilityai/sd-turbo",
        "attributor": "daam",
        "prompt": prompt,
        "ignore_special_tokens": True
    }
    for prompt in PROMPTS
]

print(f"Submitting {len(batch)} image generation jobs...")
results = client.run_smart_batch(batch, sort_strategy="none")
print("Batch completed.")
```
